Fill in the function below. The function recieves a non-empty string and an int n, and returns a new string where the character at index n has been removed.¶


```python
def remove_char(old_string, n):
    # YOUR CODE HERE
    #new_string = old_string.replace(n,"") # took in two strings but the problem said a string and an integer
    return old_string[:n] + old_string[n + 1:]

print(remove_char("hello", 1))
print(remove_char("hello", 3))
print(remove_char("hello", 4))
```

Given a string, return a new string where the first and last characters have been exchanged.

```python
def swap_chars(old_string):
  # YOUR CODE HERE
    if(old_string == ''):
        return "String too short"
        
    firstChar = old_string[0]
    lastChar = old_string[-1]
    middle = old_string[1:-1]
    new_String = ''.join(lastChar + middle + firstChar)
    return new_String

print ( swap_chars("hello") )
```

```python
# YOUR CODE HERE
print(swap_chars("BWSI"))
print(swap_chars("Medlytics"))
print(swap_chars(""))
```
